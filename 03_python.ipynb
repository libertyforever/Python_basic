### readlines() 
복수의 line 읽기

````python
f = open("io_test.txt","r")
lines = f.readlines()

for line in lines:
    line = line.strip()
    print(line)
````

## 클래스 Class
- 객체지향 언어
- 객체를 만들어내는 설계도 (붕어빵 틀)
- 클래스 안에 여러 함수나 변수를 담을 수 있음

```
class 클래스:
    클래스 내용...
```

- 통상적으로 클래스는 대문자를 사용
- self 인자 : 클래스 안에서 함수 정의시 첫번째 인자로 사용해야하는 것
 * self 를 통해 클래스 자신을 인자로 전달
 * 클래스 변수 전달 시 self.var 와 같이 작성해야만 전달 가능
- `__init__()` : 초기값을 지정하며 객체 생성
 * 
 ```py
    class CLASS1:
        def __init__(self, var1, ...)
 ```


````python
class Car:
    vel = 60
    def __init__(self,v=60):
        self.vel = v
    def velocity(self):
        return self.vel
    class Vel:
        velocity = "too high"
        def __init__(self):
            self.velocity = "slow"
````

````python
Car.vel
````

````python
car_a = Car()
````

````python
car_a.velocity()
````

````python
Car(70).vel
````

````python
Car.Vel.velocity
````

````python
Car.Vel().velocity
````

````python
class Foo:
    bar = 'a'               #클래스 전역변수
    def __init__(self):     #Foo 의 인스턴스 생성
        self.bar = 'b'      #self로 Foo 클래스 변수 전달받기
    class Bar():            #클래스 속 클래스
        bar = 'c'
        def __init__(self):
            self.bar = 'd'
````

````python
print(           
    Foo.bar,
    Foo().bar,
    Foo.Bar.bar,
    Foo.Bar().bar
    )
````
