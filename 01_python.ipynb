## 조건문


### if 조건문
```
if 조건:
    명령1
    명령2
    ...
```

- 참 거진 판단
- if문 조건 뒤는 : colon이 붙음
- 들여쓰기 4칸을 통한 명령문 구분

````python
velo = 60
if velo > 80 :
    print("속도가 {}km/h로 너무 빠릅니다.".format(velo))
elif velo > 50:
    print("속도가 정상속도 범위 입니다.".format(velo))
else:
    print("속도가 빠르지 않습니다.")
````

### 함수 리턴값

````python
def divide_f(a,b):
    div = a//b
    mod = a - div*b
    return div, mod
````

````python
divide_f(10,4)
````

### 짝수판별?

````python
def is_even(a):
    if divide_f(a,2)[1] == 0:
        return "Even"
    else:
        return "Odd"
````

````python
is_even(9)
````
