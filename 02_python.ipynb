## 반복문

### range 구문
range(end)\
range(start, end)\
range(start, end, step)

````python
for i in range(1,20,3):
    print(i, end=' ')
````

````python
sum = 0
for i in range(1,20,3):
    sum = sum + i
    print(i, end=' ')
    print(sum)
print("sum : ",sum)
````

````python
def sum_func(a,b):
    sum = 0
    for i in range(a,b+1):
        sum = sum + i
    return sum
````

````python
sum_func(10,20)
````

````python
def sum_multi(a):
    sum = 0
    for i in range(1,101):
        if i % a == 0:
            sum = sum + i
    return sum
````

````python
sum_multi(5)
````

````python
# 1에서 100 중 3 또는 5의 배수만 더하자
sum = 0
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        sum = sum + i
print(sum)
````

````python
# 사용자에게 숫자를 입력받고, 입력이 없으면 리스트로 만든 후, 만들어진 수 중 3의 배수의 합과 평균을 구하시오.

def num3multi():
    n = []
    while 1:
        num = input()
        if num == '':
            break
        n.append(int(num))
    sum = 0
    cnt = 0
    for i in n:
        if i%3==0:
            sum = sum + i
            cnt += 1
    print('총합 : ',sum)
    print('평균 : ',sum/cnt)

    
````

````python
num3multi()
````

## 집합 Set
- 수학 집합을 다루는 자료형
- 중복 순서 없음
- dictionary와 겹치는 관계로 set()로 공집합 생성
- intersection() $\rightarrow$ '&'
- union() $\rightarrow$ ' | '
- difference() $\rightarrow$ '–'

````python
set1 = {1,2,3,4,5}
print(set1)
````

````python
set2 = set()
print(set2)
````

````python
set2.add(2)
set2.add(4)
set2.add(6)
set2.add(8)
print(set2)
````

````python
print(set1 & set2)
print(set1.intersection(set2))
print(set1 | set2)
print(set1.union(set2))
print(set1 - set2)
print(set1.difference(set2))
````
